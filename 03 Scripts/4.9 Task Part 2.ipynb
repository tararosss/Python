# Step 3 - Histogram

```python
hist = instacart['order_hour_of_day'].plot.hist(bins = 24)
```

This histogram represents the peak times of day when the most orders were placed. The peak occurs around mid day which would make sense because that's probably when people think about getting groceries.  There are many users between midnight and 5 am, which would also make sense because that's when people would mostlikely be sleeping.

# Step 5

```python
#Create a random sample for data
np.random.seed(4)
```

```python
#Create dev list based on data split (70/30)
dev = np.random.rand(len(instacart)) <= 0.7
```

```python
#reduce samples to only columns needed
instacart_2 = small[['order_hour_of_day','prices']]
```

```python
instacart_2.head()
```

```python
# Create a line chart
line = sns.lineplot(data = instacart_2, x = 'order_hour_of_day',y = 'prices')
```

# Step 7

```python
# reduce previous sample to columns we need
instacart_4 = small[['age','income']]
```

```python
#create scatterplot
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = instacart_4)
```

There is a wide range of incomes in each age group.  The age groups below 40 seem to all be capped of a lower range of income, when after the age of 40 the income bracket becomes a lot higher.  This could suggest that income increases with age. The scatterplot helps us see trends in income brackets based off of certain ages.
